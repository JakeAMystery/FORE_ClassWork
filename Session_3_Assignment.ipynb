```python
def calculate(accuracy, error):
  #Returns the mean of the lowest 10 error metric
  return sum(sorted(accuracy[error])[:10])/10

def variable_hidden_layer_model(size, X, y, error):
  #size should be a list of layer densities
  shape_ = X.shape
  #print(size, shape_)
  model3 = keras.Sequential()
  model3.add(layers.Input( shape = (shape_[1],) ) )
  for i in size:
    model3.add(layers.Dense(i,'relu'))
  model3.add(layers.Dense(1,))
  model3.compile( loss = error, metrics=['mae', 'mse']) #"mae", "mse"
  history = model3.fit(X,y, epochs = 100 , verbose=0)
  pred = np.round(model3.predict(X))
  return history.history

def Accuray_comparison(layer, X, y, error):
  accuracy_dict = dict()
  for subset in car_product([10**(i) for i in range(4)], repeat=layer):
    density = tuple(subset)
    accuracy = variable_hidden_layer_model(density, X, y, error)
    accuracy_dict[density] = calculate(accuracy, error)
    print(f'{error:5}{*density,}\n{accuracy_dict[density]:.2f} Score')
  print('')
  return accuracy_dict
```

```python
def calculate(accuracy, error):
  #Returns the mean of the lowest 10 error metric
  return sum(sorted(accuracy[error])[:10])/10

def variable_hidden_layer_model(size, X, y, error, activation):
  #size should be a list of layer densities
  shape_ = X.shape
  #print(size, shape_)
  model3 = keras.Sequential()
  model3.add(layers.Input( shape = (shape_[1],) ) )
  for i in size:
    model3.add(layers.Dense(i,activation))
  model3.add(layers.Dense(1,))
  model3.compile( loss = error, metrics=['mae', 'mse']) #"mae", "mse"
  history = model3.fit(X,y, epochs = 100 , verbose=0)
  pred = np.round(model3.predict(X))
  return history.history

def Accuray_comparison(layer, X, y, error, activation):
  accuracy_dict = dict()
  for subset in car_product([10**(i) for i in range(4)], repeat=layer):
    density = tuple(subset)
    accuracy = variable_hidden_layer_model(density, X, y, error, activation)
    accuracy_dict[density] = calculate(accuracy, error)
    print(f'{error:5}{*density,}\n{accuracy_dict[density]:.2f} Score')
  print('')
  return accuracy_dict
```
